# Draw the colormaps

We use `matplotlib` to generate a RGB image of a 2D grid on the [0:1]x[0:1] subspace. The 2D vectors are converted to their RGB equivalent thanks to the 2D colormaps.


Import statements:


```python
"""Draw the colormaps."""
import colormap2d
import matplotlib.pyplot as plt
import numpy as np
```

Generation of the [0:1]x[0:1] 2D grid with 51x51 samples (the colormap data has also been sampled at 51x51):


```python
xs = np.linspace(0, 1, 51)
ys = np.linspace(0, 1, 51)

x, y = np.meshgrid(xs, ys)
grid = np.stack((x, y), axis=-1)

print(f"Shape of grid: {grid.shape}")
```

### `pinwheel` colormap:


```python
cm = colormap2d.pinwheel(grid)
plt.title("Pinwheel colormap")
plt.imshow(cm, origin="lower", extent=[0, 1, 0, 1])
```

### `cyclic_pinwheel` colormap:


```python
im = colormap2d.cyclic_pinwheel(grid)
plt.title("Cyclic pinwheel colormap")
plt.imshow(im, origin="lower", extent=[0, 1, 0, 1])
```
